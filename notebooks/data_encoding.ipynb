```python
import pandas as pd

# Import the datasets
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")

# Removes the outcome column from the dataframe
target_column = "PitNextLap"
y = train[target_column]
train_features = train.drop(columns=[target_column])

# Combines both Train and Test to be modified
# concat combines [a and b] and the axis = 0 means it stacks them vertically by rows
combined = pd.concat([train_features, test], axis=0)

# One hot encoding 
combined_encoded = pd.get_dummies(combined, drop_first=False)

# Seperation
# Sets the encoded version of train dataframe using integer location indexing from the start to the length value of train_features
train_encoded = combined_encoded.iloc[:len(train_features)]
# Sets the encoded version of test dataframe using the iloc [length value : end]
test_encoded=  combined_encoded.iloc[len(train_features):]

print(train_encoded.shape)
print(test_encoded.shape)

# Checking number of unique values each categorical column has
train.select_dtypes(include=["object"]).nunique().sort_values(ascending=False)
```
