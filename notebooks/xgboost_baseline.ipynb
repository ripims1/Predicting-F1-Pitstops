```python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent))

from src.utils import *

import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

from xgboost import XGBClassifier
```

```python
# Loading dataframes
x, y, x_test = load_processed_data()

# Train Test Split from sklearn
x_train, x_val, y_train, y_val = train_test_split(
    x,
    y,
    test_size = 0.2,
    random_state = 42,
    stratify = y
)

# Model defintion
model = XGBClassifier(
    n_estimators = 1000,
    learning_rate = 0.05,
    max_depth = 6,
    objective = "binary:logistic",
    eval_metric = "auc",
    random_state = 42
)

# Training model
model.fit(
    x_train,
    y_train,
    eval_set = [(x_val,y_val)]
 )

# Validation prediction
validaiton_prediction = generate_probability_predict(model, x_val)

# Generate estimated auc value
generate_auc(y_val, validaiton_prediction)

# Test predictions
predictions = generate_probability_predict(model, x_test)

# Generate submission
generate_submission(predictions, "../submissions/xgboost_baseline_sub.csv")
```
